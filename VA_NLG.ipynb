```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
from torch.utils.data import Dataset, DataLoader
import pandas as pd
import time

import torch.nn.functional as F
from torch.nn.utils.rnn import pad_sequence

class StoryDataset(Dataset):
    def __init__(self, texts, tokenizer, max_length=None):
        self.texts = texts
        self.tokenizer = tokenizer
        # if max_length is None:
        #     # Set max_length to the maximum sequence length in the dataset
        #     self.max_length = max(len(tokenizer.encode(text)) for text in texts)
        # else:
        self.max_length = max_length

    def __len__(self):
        return len(self.texts)

    # def __getitem__(self, idx):
    #     text = self.texts[idx]
    #     inputs = self.tokenizer.encode(text, add_special_tokens=False, truncation=True)
    #     if self.max_length:
    #         if len(inputs) >= self.max_length:
    #             inputs = inputs[:self.max_length]
    #         else:
    #             inputs += [self.tokenizer.pad_token_id] * (self.max_length - len(inputs))
    #     if inputs is None:
    #       print(inputs)
    #       inputs = []
    #     return torch.tensor(inputs, dtype=torch.long)
    def __getitem__(self, idx):
      text = self.texts[idx]
      inputs = self.tokenizer.encode(text, add_special_tokens=False, truncation=True)
      if self.max_length:
          if len(inputs) >= self.max_length:
              inputs = inputs[:self.max_length]
          else:
              inputs += [self.tokenizer.pad_token_id] * (self.max_length - len(inputs))
      inputs = [token if token is not None else 0 for token in inputs]
      words = text.split()
      return torch.tensor(inputs, dtype=torch.long)



    # def collate_fn(self, batch):
    #     return pad_sequence(batch, batch_first=True, padding_value=self.tokenizer.pad_token_id)

# Load your dataset
data = dataset.copy()
texts = data['text'].tolist()

# Initialize tokenizer and model
tokenizer = GPT2Tokenizer.from_pretrained('gpt2-medium')
model = GPT2LMHeadModel.from_pretrained('gpt2-medium')

# Define dataset and dataloader
max_length = 200
tokinize_dataset = StoryDataset(texts, tokenizer, max_length)
dataloader = DataLoader(tokinize_dataset, batch_size=2, shuffle=True)
num_epochs = 10

# Fine-tune the model
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)
model.train()

optimizer = torch.optim.AdamW(model.parameters(), lr=5e-5)

for epoch in range(num_epochs):
    start_time = time.time()
    total_loss = 0
    for i, batch in enumerate(dataloader):
        batch = batch.to(device)
        outputs = model(input_ids=batch, labels=batch)
        loss = outputs.loss
        total_loss += loss.item()


        # Update parameters
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()

        # Print batch progress
        print(f'Epoch [{epoch+1}/{num_epochs}], Batch [{i+1}/{len(dataloader)}], Loss: {loss.item():.4f}')

    # Calculate time elapsed for the epoch
    elapsed_time = time.time() - start_time
    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {total_loss/len(dataloader):.4f}, Time: {elapsed_time:.2f} seconds')
output_dir = "/content/drive/My Drive/Colab Notebooks/Pramod/models/fine_tuned_gpt2_model_10epochs_500_texts_200_words"
model.save_pretrained(output_dir)
# Generate stories based on character names and genre prompt
def generate_story(character_name, genre, max_length=100):
    prompt = f"Once upon a time, {character_name} walked into a {genre} setting and"
    input_ids = tokenizer.encode(prompt, return_tensors='pt').to(device)

    output = model.generate(input_ids, max_length=max_length, temperature=0.7, num_return_sequences=1)
    story = tokenizer.decode(output[0], skip_special_tokens=True)
    return story

# Example usage
character_name = "Alice"
genre = "fantasy"
generated_story = generate_story(character_name, genre)
print(generated_story)
```
